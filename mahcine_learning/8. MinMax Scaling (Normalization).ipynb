```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler

X, y = load_iris(as_frame=True, return_X_y=True)

```

```python
import seaborn as sns
sns.kdeplot(data=X, x='sepal length (cm)')
```

```python
print(X['sepal length (cm)'].min())
X['sepal length (cm)'].max()
```

```python
X['minus_min'] = X['sepal length (cm)'] - X['sepal length (cm)'].min()

sns.kdeplot(data=X, x='minus_min')
```

```python
X['normalized'] = X['minus_min'] / (X['sepal length (cm)'].max() - X['sepal length (cm)'].min())
print(X['sepal length (cm)'].max() - X['sepal length (cm)'].min())
sns.kdeplot(data=X, x='normalized')
```

```python
mscaler = MinMaxScaler()
X_out = pd.DataFrame()
X_out[["sepal length (cm)", "sepal width (cm)"]] = mscaler.fit_transform(X[["sepal length (cm)", "sepal width (cm)"]])
```

```python

sns.kdeplot(data=X_out, x='sepal length (cm)')
```

```python
sns.kdeplot(data=X_out, x='sepal width (cm)')
```
