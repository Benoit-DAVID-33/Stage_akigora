```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.compose import ColumnTransformer

X, y = load_iris(as_frame=True, return_X_y=True)

```

```python
import seaborn as sns
sns.kdeplot(data=X, x='sepal length (cm)')
```

```python
print(X['sepal length (cm)'].mean())
X['sepal length (cm)'].std()
```

```python
X['centered'] = X['sepal length (cm)'] - X['sepal length (cm)'].mean()

sns.kdeplot(data=X, x='centered')
```

```python
X['standardized'] = X['centered'] / X['centered'].std()
sns.kdeplot(data=X, x='standardized')
```

```python
sscaler = StandardScaler()
X_out = pd.DataFrame()
X_out[["sepal length (cm)", "sepal width (cm)"]] = sscaler.fit_transform(X[["sepal length (cm)", "sepal width (cm)"]])
```

```python

sns.kdeplot(data=X_out, x='sepal length (cm)')
```

```python
sns.kdeplot(data=X_out, x='sepal width (cm)')
```
